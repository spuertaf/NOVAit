```python
import pandas as pd
import numpy as np
from porcentajes import Porcentajes
```

```python
charlas = pd.read_excel("C:\\Users\\Admin\\Desktop\\NOVAit\\NOVAtos-20231\\datos\\crudo\\charla.xlsx")
vtg = pd.read_csv("C:\\Users\\Admin\\Desktop\\NOVAit\\NOVAtos-20231\\datos\\limpio\\vtg_clean.csv")
```

```python
charlas.sample(1)
```

```python
charlas = charlas[["Correo electrónico", "Nombre", "¿En qué horario asististe?"]] #me quedo con columnas relevantes
charlas.columns = ["Correo", "Nombre", "Asistencia Charla"] #renombrando las columnas
```

```python
#inner join con vtg para saber semestre y carrera de personas que fueron a la charla
vtg_charlas = pd.merge(charlas, vtg, how="inner", on="Correo")
vtg_charlas.sample(1)
```

Communities

```python
communities = vtg_charlas[(vtg_charlas["Depto 1"] == "Communities") | (vtg_charlas["Depto 2"] == "Communities")]
```

```python
semestres = Porcentajes(communities, "Semestre", None).obtener_porcentajes()
semestres
```

```python
carreras = Porcentajes(communities, "Carrera", "Numero Personas").obtener_porcentajes()
carreras
```
