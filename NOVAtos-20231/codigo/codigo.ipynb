```python
import pandas as pd
import numpy as np
```

```python
#Leemos los Excel con los datos de entrada
vtg = pd.read_excel("./vtg.xlsx")
asessment = pd.read_excel("./assesment.xlsx")
charla = pd.read_excel("./charla.xlsx")
```

```python
#Analizando el contenido del archivo "vtg" (Inscipciones evento: Vive tus Grupos)
vtg.sample(5)
```

```python
#Limpio data de archivo VTG
vtg = vtg.drop(["ID", "Hora de inicio", "Hora de finalización", "Correo electrónico", 
                "Nombre", "Nombre de quién te registró", "¿Tienes alguna duda o inquietud? ¡Háznosla saber!"], axis = 1)
vtg.columns = ['Nombres y apellidos', 'Correo institucional', 'Pregrado', 'Semestre en curso',
                '¿Cuál de los siguientes departamentos llama tu atención?',
                'Horario charlas informativas']
vtg.sample(2)
```

```python
#Analizar contenido de archivo "charla" (Charla Informativa sobre el grupo)
charla.sample(3)
```

```python
#Limpio data de archivo charla
charla = charla.drop(["ID", "Hora de inicio", "Hora de finalización"], axis = 1)
charla.columns = ["Correo institucional", "Nombre", "AsistenciaCharla"]
charla.sample(2)
```

```python
#Analizar contenido de archivo "asessment"
asessment.sample(3)
```

```python
#Limpio columnas innecesarias de archivo asessment
asessment = asessment[["Correo electrónico", "Nombre", "¿Asistirás al Assessment?", "¿A cuál departamento estás aspirando?"]]
asessment.columns = ["Correo", "Nombre", "AsistenciaAssessment", "Departamento"]
asessment.sample(3)
```

```python
#Inner join con correo entre los diferentes DataFrames
vtgYCharlas = pd.merge(vtg, charla, on = "Correo institucional", how="left")
vtgYCharlas = vtgYCharlas.convert_dtypes()
```

```python
vtgYCharlas["AsistenciaCharla"] = vtgYCharlas["AsistenciaCharla"].fillna("No Asiste")
```
