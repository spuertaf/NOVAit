```python
#renombramiento de todas las columnas
vtg.columns = ['ID', 'Hora de inicio', 'Hora de finalización', 'Correo electrónico',
       'Nombre', 'Nombres y apellidos', 'Correo institucional', 'Pregrado',
       'Semestre en curso',
       '¿Cuál de los siguientes departamentos llama tu atención?',
       'Horario charlas informativas', 'Nombre de quién te registró',
       '¿Tienes alguna duda o inquietud? ¡Háznosla saber!']
```

```python
#slicing de DataFrame para conservar la informacion mas relevante
vtg = vtg[["Correo institucional", "Nombres y apellidos", "Pregrado", "Semestre en curso", 
           "¿Cuál de los siguientes departamentos llama tu atención?"]]

#cambio nombre de las columnas
vtg.columns = ["Correo", "Nombre", "Carrera", "Semestre", "Departamentos"]
```

```python
#inicializo columnas de opciones de departamentos con ceros
lista_ceros = np.zeros(len(vtg["Departamentos"]))
for i in ["Depto 1", "Depto 2", "Depto 3"]:
    vtg[i] = lista_ceros
```

```python
#creo funcion para creacion de columnas con departamentos posibles de pertenencia
def add_departamentos(departamentos:str, row: int) -> None:
    list_departamentos = departamentos.split(";")[:-1] #elimino ultima posicion ya que siempre es un espacio
    list_departamentos = sorted(list_departamentos) #ordeno la lista alfabeticamente
    if len(list_departamentos) == 1:
        vtg.loc[row,"Depto 1"] = list_departamentos[0]
        vtg.loc[row,"Depto 2"] = np.NAN
        vtg.loc[row,"Depto 3"] = np.NAN
    elif len(list_departamentos) == 2:
        vtg.loc[row,"Depto 1"] = list_departamentos[0]
        vtg.loc[row,"Depto 2"] = list_departamentos[1]
        vtg.loc[row,"Depto 3"] = np.NAN
    elif len(list_departamentos) == 3:
        vtg.loc[row,"Depto 1"] = list_departamentos[0]
        vtg.loc[row,"Depto 2"] = list_departamentos[1]
        vtg.loc[row,"Depto 3"] = list_departamentos[2]
```

```python
#creando columnas con departamentos posibles de pertenencia
posibles_departamentos = vtg["Departamentos"]
for i in range(0, len(posibles_departamentos)):
    add_departamentos(posibles_departamentos[i], row= i)
```

```python
#elimino columna de deparatamentos
vtg = vtg.drop(columns=["Departamentos"])
```

### Analizando la informacion

```python
def obtener_porcentajes(dataframe:pd.DataFrame, columna: str, sort:str=None) -> pd.DataFrame:
    assert columna in dataframe.columns, f"{columna} no es una columna del dataframe"
    porcentajes = dataframe[columna].value_counts().sort_index().to_frame()
    porcentajes = porcentajes.reset_index()
    porcentajes.columns = ["Semestre", "Numero Personas"]
    if sort is not None:
        porcentajes = porcentajes.sort_values(by=sort, ascending=False)
    porcentajes["Porcentajes"] = round((porcentajes["Numero Personas"] / sum(porcentajes["Numero Personas"])),3)*100
    porcentajes["Acumulado Porcentajes"] = porcentajes["Porcentajes"].cumsum()
    return porcentajes
```

Communities

```python
communities = vtg[(vtg["Depto 1"] == "Communities") | (vtg["Depto 2"] == "Communities")]
```

```python
communities.sample(1)
```

```python
semestres = obtener_porcentajes(communities, "Semestre")
semestres["Acumulado Personas"] = semestres["Numero Personas"].cumsum() #saco acumulado de personas
semestres
```

```python
carreras = obtener_porcentajes(communities, "Carrera", sort = "Numero Personas")
carreras
```
